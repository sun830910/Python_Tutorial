# 函式-基本使用
經常重覆使用的方法就寫成函式(functions).

```python
#-----------------------------
# 一個印出分隔線的函式
# 傳入:無
# 回傳:無
#-----------------------------
def separateLine():
    print('-'*30)
    print('='*30)


#-----------------------------
# 主程式
#-----------------------------
for i in range(1, 100):
    if i%3==0 or i%5==0 or i%7==0:
        separateLine()
    else:
        print(i)
```

```python
#-----------------------------
# 一個計算傭金的函式
# 傳入:銷售金額
# 回傳:傭金
#-----------------------------
def commission(total):
    rate=0.0
    
    if total>20000:
        rate=0.15
    elif total>10000:
        rate=0.1
    else:
        rate=0.08
        
    return int(total*rate)
#-----------------------------    



#-----------------------------
# 主程式
#-----------------------------
tot=[25000, 36000, 12000, 9500, 45000]

for t in tot:
    print('銷售{}元, 傭金{}元'.format(t, commission(t)))  
```

```python
#-----------------------------
# 一個計算傭金的函式
# 傳入:銷售金額, 職級
# 回傳:傭金 
#-----------------------------
def commission(total, rank):
    rate=0.0
    
    if rank=='A':
        if total>20000:
            rate=0.15
        elif total>10000:
            rate=0.1
        else:
            rate=0.08
    else:
        if total>20000:
            rate=0.12
        elif total>10000:
            rate=0.09
        else:
            rate=0.05        
        
    return int(total*rate)   
#-----------------------------



#-----------------------------
# 主程式
#-----------------------------
tot=[(25000, 'A'), (36000, 'B'), (12000, 'B'), (9500, 'A'), (45000, 'A')]

for t, r in tot:
    print('銷售{}元, 職級{}, 傭金{}元'.format(t, r,  commission(t, r))) 
#-----------------------------    
```
