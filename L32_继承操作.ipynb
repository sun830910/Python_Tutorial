# 類別(class)-繼承
在宣告類別敘述的小括號中加入另一個類別名稱, 加入的就是父類別, 自己稱為子類別. 所有類別最上層的父親都是object.

     object
       ^  
       |  
   Student類別       宣告成: class Student(): 或 class Student(object):  
       ^  
       |   
  IM_Student類別     宣告成: class IM_Student(Student):

```python
#------------------------------------------------
# 定義學生類別
#------------------------------------------------
class Student():
    #------------------------------       
    # 建構元
    # 為了重新定義gender的取存, 
    # 其實際存放名稱改為 _gender.
    #------------------------------   
    def __init__(self, stuNo, stuName, gender=None):
        self.stuNo=stuNo
        self.stuName=stuName
        self._gender=gender   #這行指令會呼叫@gender.setter方法

    #----------------------------------    
    # 取用gender屬性的方法
    # 直接用 [...=實例名稱.gender] 即可
    #---------------------------------- 
    @property
    def gender(self):
        if self._gender=='M':
            return '男'
        elif self._gender=='F':
            return '女'
        else:
            return None

    #----------------------------------    
    # 寫入gender屬性的方法
    # 直接用 [實例名稱.gender=...] 即可
    #---------------------------------- 
    @gender.setter
    def gender(self, gender):
        if gender=='M' or gender=='F':
            self._gender=gender
        else:
            self._gender=None
    

    #-------------------    
    # 取出學制的方法
    #-------------------
    def division(self):
        if self.stuNo[0]=='N':
            return '進修推廣部'
        else:
            return '日間部'
            
            

#------------------------------------------------
# 定義資管系學生類別, 它繼承了學生類別
#------------------------------------------------
class IM_Student(Student):
    #---------------------------------------     
    # 建構元
    # 用super().__init__()呼叫父親的建構元
    #---------------------------------------
    def __init__(self, stuNo, stuName, skill, gender=None):
        super().__init__(stuNo, stuName, gender)
        self.skill=skill

    #---------------------------------------      
    # 判斷學生專長種類的方法
    #---------------------------------------        
    def skillType(self):
        if self.skill=='Java' or self.skill=='Python':
            return '程式專長'
        elif self.skill=='HTML5':
            return '網頁專長'
        else:
            return '其他專長'        
```

```python

#--------------------------------------------
# 生成實例
#--------------------------------------------
s1=Student('1056001', '王小明', 'M')
s2=IM_Student('1056002', '陳小華', 'Java', 'F')

#--------------------------------------------
# 呼叫實例的方法
#-------------------------------------------- 
print('學號:{}  姓名:{}  性別:{}  學制:{}'.format(s1.stuNo, s1.stuName, s1.gender, s1.division()))
print('學號:{}  姓名:{}  性別:{}  學制:{}  專長:{}  專長種類:{}'.format(s2.stuNo, s2.stuName, s2.gender, s2.division(), s2.skill, s2.skillType()))
```

## Python支援多重繼承.

```python
#------------------------------------------------
# 定義學生類別
#------------------------------------------------
class Student(object):
    #------------------------------       
    # 建構元
    # 為了重新定義gender的取存, 
    # 其實際存放名稱改為 _gender.
    #------------------------------   
    def __init__(self, stuNo, stuName, gender=None):
        self.stuNo=stuNo
        self.stuName=stuName
        self._gender=gender   #這行指令會呼叫@gender.setter方法

    #----------------------------------    
    # 取用gender屬性的方法
    # 直接用 [...=實例名稱.gender] 即可
    #---------------------------------- 
    @property
    def gender(self):
        if self._gender=='M':
            return '男'
        elif self._gender=='F':
            return '女'
        else:
            return None

    #----------------------------------    
    # 寫入gender屬性的方法
    # 直接用 [實例名稱.gender=...] 即可
    #---------------------------------- 
    @gender.setter
    def gender(self, gender):
        if gender=='M' or gender=='F':
            self._gender=gender
        else:
            self._gender=None
    

    #-------------------    
    # 取出學制的方法
    #-------------------
    def division(self):
        if self.stuNo[0]=='N':
            return '進修推廣部'
        else:
            return '日間部'
  
            
#------------------------------------------------
# 定義專長類別
#------------------------------------------------
class Skill(object):
    #------------------------------       
    # 建構元
    #------------------------------ 
    def __init__(self, name):
        self.name=name
        
    #---------------------------------------      
    # 判斷學生專長種類的方法
    #---------------------------------------        
    def type(self):
        if self.name=='Java' or self.name=='Python':
            return '程式專長'
        elif self.name=='HTML5':
            return '網頁專長'
        else:
            return '其他專長'        
        

#------------------------------------------------
# 定義資管系學生類別, 它繼續了學生及專長類別
#------------------------------------------------
class IM_Student(Student, Skill):
    #---------------------------------------     
    # 建構元
    # 用Student.__init__()及Skill.__init__()
    # 分別呼叫2個父親的建構元
    #---------------------------------------
    def __init__(self, stuNo, stuName, skill, gender=None):
        Student.__init__(self, stuNo, stuName, gender)
        Skill.__init__(self, skill)
```

```python
#--------------------------------------------
# 生成實例
#--------------------------------------------
s1=Student('1056001', '王小明', 'M')
s2=IM_Student('1056002', '陳小華', 'Java', 'F')

#--------------------------------------------
# 呼叫實例的方法
#-------------------------------------------- 
print('學號:{}  姓名:{}  性別:{}  學制:{}'.format(s1.stuNo, s1.stuName, s1.gender, s1.division()))
print('學號:{}  姓名:{}  性別:{}  學制:{}  專長:{}  專長種類:{}'.format(s2.stuNo, s2.stuName, s2.gender, s2.division(), s2.name, s2.type()))
```
