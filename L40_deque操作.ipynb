# 資料型態(Data types)-deque
deque是一個頭尾兩端都可以增刪資料的結構, 如果有必要, 也可以從頭尾之間插入資料.

```python
#--------------------------------------------------
# 匯入collections模組中的deque類別
#--------------------------------------------------
from collections import deque

#--------------------------------------------------
# 產生一個deque實例
#--------------------------------------------------
dq=deque()

#--------------------------------------------------
# append(data)及appendleft(data)
#--------------------------------------------------
dq.append(10)
dq.appendleft(20)
print('動作: append(10)及appendleft(10)')
print('結果:', dq)
print('-'*30)

#--------------------------------------------------
# insert(index, data)
#--------------------------------------------------
dq.insert(1, 30)
print('動作: insert(1, 30)')
print('結果:', dq)
print('-'*30)

#--------------------------------------------------
# pop()及popleft()
#--------------------------------------------------
print('動作: pop()')
print(dq.pop())
print('結果:', dq)
print('-'*30)

print('動作: popleft()')
print(dq.popleft())
print('結果:', dq)
print('-'*30)

#--------------------------------------------------
# extend(data)及extendleft(data)
#--------------------------------------------------
dq.extend([100, 200, 300, 400, 500, 100])
print('動作: extend([100, 200, 300, 400, 500, 100])')
print('結果:', dq)
print('-'*30)

#--------------------------------------------------
# remove(data), 移除第一個符合的元素
#--------------------------------------------------
dq.remove(100)
print('動作: remove(100)')
print('結果:', dq)
print('-'*30)

#--------------------------------------------------
# reverse(), 反轉
#--------------------------------------------------
dq.reverse()
print('動作: reverse()')
print('結果:', dq)
print('-'*30)

#--------------------------------------------------
# reverse(n), 向右旋轉2次
#--------------------------------------------------
dq.rotate(2)
print('動作: rotate(2)')
print('結果:', dq)
print('-'*30)

#--------------------------------------------------
# 傾印deque內容
#--------------------------------------------------
for d in dq:
    print(d)
print('-'*30)
```
